```
# Read the file that contain Title token and Description token
df_train_X = pd.read_csv("gdrive/My Drive/5500/job-salary-prediction/Train_X_modified.csv")
```

```
# Transform the list of of text tokens to text used in tf-idf vectorizer
# Reason not using CountVector from sklearn: want to be uniformed with bag of words method
def list_to_count_dic(Text_List):
  new_text_list = []
  for i in tqdm(Text_List):
    temp_str = ""
    spited_i=i[1:-1].split(', ')
    for j in spited_i:
      if j[1:-1] != ' ':
        temp_str = temp_str + j[1:-1] + " "
    new_text_list.append(temp_str[:-1])
  
  return new_text_list
```

```
# Use TfidfVectorizer from sklearn to transform text to vectors
def tf_idf(Text_List, max_feature):
    vectorizer = TfidfVectorizer(max_features = max_feature, ngram_range=(1,2))
    X = vectorizer.fit_transform(list_to_count_dic(Text_List))
    print(vectorizer.get_feature_names())
    return X.toarray()
```
